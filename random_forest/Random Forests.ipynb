### Import the required libraries

```python
import numpy as np #
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split  
from sklearn.ensemble import RandomForestClassifier 

import warnings
warnings.filterwarnings('ignore')
```

### Load the data

```python
data = 'data_randomforest.csv'  # this data is about cars features 

df = pd.read_csv(data, header=None)  # to skip the header raw
df.head()
```

### Explore the data

```python
# view dimensions of the dataset
df.shape

# after running the cell, we can see that there are 1728 instances and 7 variables in the data set.
```

```python
# rename the variables of the dataset
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
df.columns = col_names
df.head()
```

```python
# get a summary of the dataset
df.info()
```

### Preprocess the data

```python
# check missing values in variables
df.isnull().sum()
```

```python
# determine the dependent and independent variables
X = df.drop(['class'], axis=1)
y = df['class']
```

```python
# split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# check the shape of X_train and X_test
X_train.shape, X_test.shape
```

```python
# import category encoders
import category_encoders as ce
```

```python
# encode variables with ordinal encoding
encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
# let´s see how X_train looks like now
X_train.head()
```

### Build the model

```python
# instantiate the classifier 
rfc = RandomForestClassifier(random_state=0)

# fit the model
rfc.fit(X_train, y_train)
```

```python
# Predict the Test set results
y_pred = rfc.predict(X_test)
```

#### The model accuracy

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
